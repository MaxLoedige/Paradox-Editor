```python
# Modeled after a rudementary commmon syntax parser but there is still a lot missing

def _parseBlock(text, start=0):
    scope = {}
    key = ""
    value = ""
    in_key = True
    comment = False
    i = start

    inline = 0

    while i < len(text):
        char = text[i]

        # Handle comments
        if char == "#":
            comment = True
        if comment:
            if char == "\n":
                comment = False
            i += 1
            continue

        # End of block
        if char == "}":
            if key.strip():
                scope[key.strip()] = value.strip()
            return scope, i

        # Start of nested block
        if char == "{":
            sub_key = key.strip().rstrip("=")
            sub_block, new_i = _parseBlock(text, i + 1)
            scope[sub_key] = sub_block
            key = ""
            value = ""
            in_key = True

            # inline state is reset here
            inline = 0
            
            i = new_i + 1  # Resume after nested block
            continue

        # Handle newlines as entry delimiters
        if char == "\n":
            if key.strip():
                scope[key.strip().rstrip("=")] = value.strip()
            key = ""
            value = ""
            in_key = True
            

            inline = 0

            i += 1
            continue

        # Handle '=' character to switch from key to value
        if char in ["=","<",">","!"] and in_key:
            in_key = False
            inline = 1
            i += 1
            if text[i] == "=":
                i += 1
            continue
        
        if inline == 1 and char != " ":
            inline = 2
        if inline == 2 and char == " ":
            if text[i+1].isalpha():
                inline = 3
                continue
        if inline == 3:
            if key.strip():
                scope[key.strip().rstrip("=")] = value.strip()
            key = ""
            value = ""
            in_key = True
            

            inline = 0

            i += 1
            continue


        if in_key:
            key += char
        else:
            value += char

        i += 1

    return scope, i
```

```python
p = Path(r"D:\Programmiren\ck3 modding refferences\sinews-of-war-main\main\common\character_interactions")
character_interactions = {}
for f in p.glob("*.txt"):
    with open(f , "rb") as file:
        data = file.read().decode("utf-8-sig")
        t,_ = _parseBlock(data)
        character_interactions.update(t)
```

```python
from collections import Counter

# Your big dictionary
interactions = character_interactions

# Extract all first-level keys
all_keys_per_entry = [set(data.keys()) for data in interactions.values()]

# Count key frequency
key_counter = Counter()
for keys in all_keys_per_entry:
    key_counter.update(keys)

num_entries = len(interactions)

# Classify keys
common_keys = {key for key, count in key_counter.items() if count == num_entries}
optional_keys = {key for key, count in key_counter.items() if count < num_entries}
required_keys = common_keys  # optional naming

# Output
print("Common (Required) Keys:", sorted(common_keys))
print("Optional Keys:", sorted(optional_keys))
```
